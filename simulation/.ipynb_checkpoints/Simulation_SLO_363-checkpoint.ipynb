```python
infer_cnt=0
infer_cnt_list=[]
train_cnt=0
train_cnt_list=[]
first_time = start_time_list[0]
day_cnt = 1
util_train = []
util_infer = []
util_train_day = []
util_infer_day = []
mem_train = []
mem_infer = []
mem_train_day = []
mem_infer_day = []
duration_train = []
duration_infer = []
thresh_t = 200000 #50000:infer:train=3:1, 14000:infer:train=2:1, 3000:infer:train=1:1, 1250:infer:train=0.5:1
thresh_u = 20 # 20
for i,u in enumerate(duration_list):
    if (start_time_list[i]-first_time)/3600 <= day_cnt:
        if (u > thresh_t):
            train_cnt=train_cnt+1
            util_train.append(util_list[i])
            mem_train.append(mem_list[i])
            duration_train.append(u)
        else:
            if util_list[i] > thresh_u:
                train_cnt=train_cnt+1
                util_train.append(util_list[i])
                mem_train.append(mem_list[i])
                duration_train.append(u)
            else:
                infer_cnt=infer_cnt+1
                util_infer.append(util_list[i])
                mem_infer.append(mem_list[i])
                duration_infer.append(u)
    else:
        train_cnt_list.append(train_cnt)
        infer_cnt_list.append(infer_cnt)
        train_cnt=0
        infer_cnt=0
        if (u <= (thresh_t)) and (util_list[i] <= thresh_u): 
            infer_cnt=infer_cnt+1
        else:
            train_cnt=train_cnt+1
            util_train.append(util_list[i])            
        util_train_day.append(np.mean(util_train))
        util_train.clear()
        util_infer_day.append(np.mean(util_infer))
        util_infer.clear()
        mem_train_day.append(np.mean(mem_train))
        mem_train.clear()
        mem_infer_day.append(np.mean(mem_infer))
        mem_infer.clear()
        day_cnt = day_cnt+1
fig, ax = plt.subplots()
fig.set_size_inches(4.5, 3)
ax.plot(np.arange(0,len(train_cnt_list)),train_cnt_list, label='Train')
ax.plot(np.arange(0,len(train_cnt_list)),infer_cnt_list, label='Infer')
ax.legend()
# plt.xlim([21, 24])
ax.set_xlabel('Hours')
ax.set_ylabel('Number of Jobs')
```
